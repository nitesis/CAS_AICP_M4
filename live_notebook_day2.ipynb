```python
import numpy as np
import skimage
import matplotlib.pyplot as plt
```

```python
im_large = np.zeros((1000,1000, 3), dtype=np.uint8)
```

```python
rr, cc = skimage.draw.disk(center=(400, 600), radius=50, shape=(1000, 1000))
```

```python
cc
```

```python
im_large[rr, cc, 0] = 255
```

```python
# Change the color of the image at the coordinates (rr, cc) to the defined color in RGB format.
skimage.draw.set_color(image = im_large, coords=(rr, cc), color=(200, 100, 100))
```

```python
im_large = np.zeros((1000, 1000, 3), dtype=np.uint8)
# Adding 4 differnet positions  
row_pos = [400, 100, 900, 300]
col_pos = [800, 400, 330, 102]

for i in range (4):
    row, col = skimage.draw.disk(center=(row_pos[i], col_pos[i]), radius=50, shape=(1000,1000))
    skimage.draw.set_color(image=im_large, coords=(row,col), color=(200, 100, 100))
```

```python
# Intitializing the variable (Note: If not, circles will added every execution on top to the existing on)
im_large = np.zeros((1000, 1000, 3), dtype=np.uint8)
# Adding random positions
row_pos = np.random.randint(low=0, high=1000, size=4)
col_pos = np.random.randint(low=0, high=1000, size=4)

for i in range (4):
    row, col = skimage.draw.disk(center=(row_pos[i], col_pos[i]), radius=50, shape=(1000,1000))
    skimage.draw.set_color(image=im_large, coords=(row,col), color=(200, 100, 100))
```

```python
# Intitializing the variable
im_large = np.zeros((1000, 1000, 3), dtype=np.uint8)
# Adding random positions
row_pos = np.random.randint(low=0, high=1000, size=4)
col_pos = np.random.randint(low=0, high=1000, size=4)

# Adding random color on the loop   
for i in range (4):
    # Creates a random color in the RGB space
    #color = np.random.randint(low=0, high=255, size=3)
    
    # Creates a random color by only changing the blue-channel randomly
    color = (255, 0, np.random.randint(low=0, high=255))
    myradius = np.random.randint(low=10, high=100)

    row, col = skimage.draw.disk(center=(row_pos[i], col_pos[i]), radius=myradius, shape=(1000,1000))
    skimage.draw.set_color(image=im_large, coords=(row,col), color=color)
```

```python
plt.imshow(im_large)
```

```python
# Create circles where the psoition depends on the position
im_large = np.zeros((1000, 1000, 3), dtype=np.uint8)
   
for i in range (1000):
    myradius = np.random.randint(low=10, high=100)
    #Setting limit on a row and transform it in an integer number
    max_row = int(10000 / myradius)

    row_pos = np.random.randint(low=0, high=max_row)
    col_pos = np.random.randint(low=0, high=1000)
    
    # Creates a random color by only changing the blue-channel randomly
    color = (255, 0, np.random.randint(low=0, high=255))
    

    row, col = skimage.draw.disk(center=(row_pos, col_pos), radius=myradius, shape=(1000,1000))
    skimage.draw.set_color(image=im_large, coords=(row,col), color=color)

    plt.imshow(im_large)
```

```python
from skimage import io

# Then save the image
io.imsave('pink_ballons.png', im_large)
```
