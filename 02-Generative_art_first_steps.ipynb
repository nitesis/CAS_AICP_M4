# Generative Art: very first steps with 

Until now we have only modified existing images but not really generated new ones "from nothing". We explore here how we can create images based on geometry and randomness with scikit-image. That package is not commonly used for generative art (more on this later) but allows us to explain concepts within a relatively simple library.

## Simple shape

First let's create an empty image using the ```np.zeros``` function which essentially creates a blank image:

````python
import numpy as np
import matplotlib.pyplot as plt
import skimage
````

````python
np.set_printoptions(linewidth=200)
myimage = np.zeros((30,30))
myimage
````

We display again the image using matplotlib and of course get a blank image:

````python
plt.imshow(myimage, cmap='gray');
````

Now we can generate coordinates for example for a disk. We need to provide:
- the disc location in terms of row/column
- the disc radius
- the size of the image in which we want to embedd that disk

````python
my_disk = skimage.draw.disk(center=[12,10], radius=8, shape=[30,30])
````

````python
my_disk
````

The output is two series of coordinates of pixels belonging to the disk: one list for rows and one for columns. Remember that we **can use these coordinates to get or set pixel values**. Let's do that and set those pixels to the value 1:

````python
myimage[my_disk] = 1
````

````python
myimage
````

````python
plt.imshow(myimage);
````

Other non-generic shapes are just defined as polygons with all their vertex coordinates:

````python
myimage = np.zeros((30,30))

polygon = skimage.draw.polygon([5,15,20,8], [8,6,15,14], shape=myimage.shape)

myimage[polygon] = 1

plt.imshow(myimage);
````

## Adding colour

We managed to create a picture, but now it's just a black and white picture. We could set a colormap but then it will be difficult to control the exact colour of objects. We will rather create now a natural RGB image, i.e. an image composed of three channels. We also set it to be an 8bit image, like natural images:

````python
im_RGB = np.zeros((30,30,3), dtype=np.uint8)
````

Now instead of directly accessing the pixels and setting certain coordinates to 1, we can use the ```set_color``` function to just color those coordinates in a certain way:

````python
skimage.draw.set_color(image=im_RGB, coords=my_disk, color=[255,0,0])
````

````python
plt.imshow(im_RGB);
````

## Generative Art

Until now, we have only "designed" some drawing. If we want to do algorithmic or generative images, we need to introduce some automation and some randomness to the process. We'll keep with our simple example of disk to explain this. From here on, we'll use a larger image to make it more interesting:

````python
im_large = np.zeros((1000,1000,3), dtype=np.uint8)
````

The first thing that we can do is add more than one disk and select it's position in a random way. We can use the ```np.random``` module to genreate such positions. For example:

````python
np.random.randint(low=0, high=1000)
````

generates a coordinate within our image boundaries. So we can do now a for loop and add three disks to our image:

````python
np.random.seed(42)
im_large = np.zeros((1000,1000,3), dtype=np.uint8)

for i in range(3):

    random_row = np.random.randint(low=0, high=1000)
    random_col = np.random.randint(low=0, high=1000)
    
    my_disk = skimage.draw.disk(center=[random_row, random_col], radius=30, shape=[1000,1000])
    
    skimage.draw.set_color(im_large, my_disk, color=[255,0,0])

plt.imshow(im_large);
````

Now we have it: our first generative art masterpiece!

## More choices

We can decide on more than just the position of our disks: we can set their radius as well as their color using the same principle. Note that to define the colour, we need three numbers, so we can directly use:

````python
np.random.randint(low=0, high=255, size=3)
````

````python
im_large = np.zeros((1000,1000,3), dtype=np.uint8)

for i in range(10):

    random_row = np.random.randint(low=0, high=1000)
    random_col = np.random.randint(low=0, high=1000)

    random_radius = np.random.randint(low=10, high=100)

    random_color = np.random.randint(low=0, high=255, size=3)
    
    my_disk = skimage.draw.disk(center=[random_row, random_col], radius=random_radius, shape=[1000,1000])
    
    skimage.draw.set_color(im_large, my_disk, color=random_color)

plt.imshow(im_large);
````

## More rules

Generative art lives of the tension between carefully crafted rules and randomness. Currently we have replaced a simple rule (set a location) with a completely random rule (generate a location). We can now try to find an interesting balance between these two by defininig a more interesting algorithm. For example, we could make the y location (row) dependent on the disk radius. How can we turn that into code. First we pick a random radius:

````python
random_radius = np.random.randint(low=10, high=100)
random_radius
````

And now we make the maximum allow row index inversely proportional to the radius i.e. $row_{max} \propto 1/r$. However we have to be careful when crafting rules:

````python
1 / random_radius
````

The current rule generates values smaller than one that we can't use as an index. The value will always be smaller than 1, so we could safely multiply this by the largest possible row: large radii will still generate smaller values but now in a reasonable range. We also have to be careful to transform this value into an integer as expected by the ```np.random.randint``` function:

````python
int(1000 * (1 / random_radius))
````

Let's try to use this rule in our routine! We will use here many more disks to not depend to much on variability between runs:

````python
im_large = np.zeros((1000,1000,3), dtype=np.uint8)

for i in range(200):

    random_radius = np.random.randint(low=10, high=100)

    # new condition here
    max_row = int(1000 * (1 / random_radius))
    random_row = np.random.randint(low=0, high=max_row)
    
    random_col = np.random.randint(low=0, high=1000)

    random_color = np.random.randint(low=0, high=255, size=3)
    
    my_disk = skimage.draw.disk(center=[random_row, random_col], radius=random_radius, shape=[1000,1000])
    
    skimage.draw.set_color(im_large, my_disk, color=random_color)

plt.imshow(im_large);
````

So now the rows seem indeed more limited but we went too far: even small disks are at the top of our image. The reason for this is that our smallest possible radius is 10, which means our largest possible row is: 

````python
1000 * (1 / 10)
````

still a small row. If we change our equation to $1000 * (10 / r)$ we are still within reasonble boundaries:

````python
im_large = np.zeros((1000,1000,3), dtype=np.uint8)

for i in range(200):

    random_radius = np.random.randint(low=10, high=100)

    # new condition here
    max_row = int(1000 * (10 / random_radius))
    random_row = np.random.randint(low=0, high=max_row)
    
    random_col = np.random.randint(low=0, high=1000)

    random_color = np.random.randint(low=0, high=255, size=3)
    
    my_disk = skimage.draw.disk(center=[random_row, random_col], radius=random_radius, shape=[1000,1000])
    
    skimage.draw.set_color(im_large, my_disk, color=random_color)

plt.imshow(im_large);
````

So now we get the desired effecf: it looks like large disks float at the top while small ones are sinking. We could also flip the array to make small disks look like the fly away.

````python
plt.imshow(np.flip(im_large, axis=0));
````

````python
skimage.io.imsave('output/balloons.png', np.flip(im_large, axis=0))
````

## chatGPT solution

Largue Language Models (LLMs) are increasingly capable of writing code in various languages in a very efficient way. Instead of having first to learn an entire syntax, one can just prompt for some script, try it out, learn from it, and modify it, then adjust again the prompt etc. allowing for a very fast iterative creation. For example here we prompted chatGPT with: *"Using scikit-image write a short script that will add multiple shapes (square, elipse etc.) to an image and give them shades of greeen."* and here's the code and output:

````python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import rectangle, ellipse
from skimage import img_as_ubyte

# Create a blank image
image = np.zeros((200, 200, 3), dtype=np.uint8)

# Define colors in shades of green
colors = [
    (0, 255, 0),     # Bright green
    (0, 155, 0),     # Medium green
    (0, 100, 0)      # Dark green
]

# Draw a square
start = (50, 50)  # Top-left corner of the square
extent = (50, 50)  # Height and width of the square
rr, cc = rectangle(start, extent=extent)
image[rr, cc] = colors[0]

# Draw an ellipse
r_center, c_center = 120, 120  # Center of the ellipse
r_radius, c_radius = 30, 20    # Radii of the ellipse
rr, cc = ellipse(r_center, c_center, r_radius, c_radius)
image[rr, cc] = colors[1]

# Add another square
start = (10, 10)
extent = (30, 30)
rr, cc = rectangle(start, extent=extent)
image[rr, cc] = colors[2]

# Convert to uint8 and display the image
image = img_as_ubyte(image)
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
````

Here's an image generated with a similar approach to what we have done above. Observe it and try to reproduce it!

````python
# Je kleiner die Kreise, um so nher zum Zentrum 
import matplotlib.pyplot as plt
import numpy as np

# Anzahl der Kreise
num_circles = 200
saturation_start = 0.4  # Entfernung vom Zentrum, ab der die Entsttigung beginnt

# Setze den Bildbereich und den Radius
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
fig.patch.set_facecolor('black')  # Hintergrundfarbe auf Schwarz setzen
ax.set_facecolor('black')

# Reduziere die Sichtachse, um Kreise ber den Bildrand hinaus zu zeichnen
ax.set_xlim(-0.8, 0.8)
ax.set_ylim(-0.8, 0.8)

# Entferne den Rahmen (Randlinien) um den Plot
for spine in ax.spines.values():
    spine.set_visible(False)

# Entferne die Achsenmarkierungen
ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)

# Erstelle die Kreise
for _ in range(num_circles):
    # Berechne zufllige Position in Polarkoordinaten
    angle = np.random.uniform(0, 2 * np.pi)
    distance = np.random.uniform(0, 1)
    
    # Je weiter vom Zentrum entfernt, desto grer und weniger gesttigt der Kreis
    radius = distance * 0.15  # Maximale Gre von Kreisen nach auen hin
    saturation = 1 - distance  # Sttigung nimmt mit der Entfernung ab

    # Sttigung anpassen: innerhalb von saturation_start volle Sttigung, danach graduell abnehmend
    if distance < saturation_start:
        saturation = 1  # Volle Sttigung
    else:
        saturation = 1 - (distance - saturation_start) / (1 - saturation_start)  # Lineare Abnahme ab saturation_start
    
    # Zufllige Farbe und Sttigung anwenden
    color = np.random.rand(3,) * saturation
    
    # Berechne die x- und y-Koordinaten
    x = distance * np.cos(angle)
    y = distance * np.sin(angle)
    
    # Zeichne den Kreis
    circle = plt.Circle((x, y), radius, color=color, alpha=0.6)
    ax.add_patch(circle)

# Zeige das Bild
plt.axis('off')
plt.show()
````
