## 2. Loading the data and more image processing

```python
im_cats = iio.imread('data/woody_bulle.JPG')
plt.imshow(im_cats)
```

```python
im_cats.shape # RGB

# Gibt nur den ersten Kanal zurück
im_R = im_cats[:,:,0]
plt.imshow(im_R)
```

```python
plt.imshow(im_R, cmap='gray')
```

```python
# Hier das Bild ausgeschnitten y-Achse 1000-2000, x-Achse 2000-3000
im_crop_01 = im_cats[1000:2000, 2000:3000, :]

print(im_crop_01.shape)   

plt.imshow(im_crop_01)
```

```python
# Hier das Bild ausgeschnitten y-Achse, x-Achse, Farbkanäle

im_crop_02 = im_cats[1000:2000, 0:1000, :]
 
print(im_crop_02.shape)

plt.imshow(im_crop_02)
```

```python
sum_crop = im_crop_01 + im_crop_02
plt.imshow(sum_crop)
```

```python
img_rotated = skimage.transform.rotate(sum_crop, 45)
plt.imshow(img_rotated)
```

## 3. Filters

```python
img_soble = skimage.filters.sobel(img_roisin)
plt.imshow(img_soble)

```

```python
img_gaussian_01 = skimage.filters.gaussian(img_roisin, preserve_range=False)
plt.imshow(img_gaussian_01)
```

```python
img_gaussian_02 = skimage.filters.gaussian(img_roisin, sigma=10)
plt.imshow(img_gaussian_02)
```

```python
img_gaussian_01
#skimage.io.imsave(img_roisin_blurry)
```

```python

img_gaussian_01 = 255*((img_gaussian_01 - img_gaussian_01.min())/(img_gaussian_01.max() - img_gaussian_01.min()))
img_gaussian_01_int = img_gaussian_01.astype(np.uint8)
skimage.io.imsave('img_gaussian_01_int.png', img_gaussian_01_int)
img_gaussian_01_int.shape
plt.imshow(img_gaussian_01_int)
```
